```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# extract data from csv
baseline = pd.read_csv('square/xy_baseline.csv')
post = pd.read_csv('square/xy_post.csv')
```

```python
baseline.head()

# caluclate mean values
a1_x, a1_y = baseline['A_x'][:5].mean(), baseline['A_y'][:5].mean()
a2_x, a2_y = baseline['A_x'][5:].mean(), baseline['A_y'][5:].mean()
b1_x, b1_y = baseline['B_x'][:5].mean(), baseline['B_y'][:5].mean()
b2_x, b2_y = baseline['B_x'][5:].mean(), baseline['B_y'][5:].mean()
c1_x, c1_y = baseline['C_x'][:5].mean(), baseline['C_y'][:5].mean()
c2_x, c2_y = baseline['C_x'][5:].mean(), baseline['C_y'][5:].mean()

# overall mean values
a_x, a_y = baseline['A_x'].mean(), baseline['A_y'].mean()
b_x, b_y = baseline['B_x'].mean(), baseline['B_y'].mean()
c_x, c_y = baseline['C_x'].mean(), baseline['C_y'].mean()

# caluclate mean values for post
pa1_x, pa1_y = post['A_x'][:5].mean(), post['A_y'][:5].mean()
pa2_x, pa2_y = post['A_x'][5:].mean(), post['A_y'][5:].mean()
pb1_x, pb1_y = post['B_x'][:5].mean(), post['B_y'][:5].mean()
pb2_x, pb2_y = post['B_x'][5:].mean(), post['B_y'][5:].mean()
pc1_x, pc1_y = post['C_x'][:5].mean(), post['C_y'][:5].mean()
pc2_x, pc2_y = post['C_x'][5:].mean(), post['C_y'][5:].mean()

# overall mean values for post
pa_x, pa_y = post['A_x'].mean(), post['A_y'].mean()
pb_x, pb_y = post['B_x'].mean(), post['B_y'].mean()
pc_x, pc_y = post['C_x'].mean(), post['C_y'].mean()
```

```python
# plot the data in a scatter plot of [A_x, A_y] and [B_x, B_y] and [C_x, C_y]
# for both the baseline and post calibration data
# make the first 5 one shape and the next 5 another shape
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# add a grid
ax[0].grid()
ax[1].grid()

# baseline opacity = 0.5
ax[0].scatter(baseline['A_x'][:5], baseline['A_y'][:5], label='CW', s=50, marker='x', c='r', alpha=0.2)
ax[0].scatter(baseline['A_x'][5:], baseline['A_y'][5:], label='CCW', s=50, marker='o', c='r',alpha=0.2)
ax[0].scatter(baseline['B_x'][:5], baseline['B_y'][:5], label='CW',s=50, marker='x', c='g',alpha=0.2)
ax[0].scatter(baseline['B_x'][5:], baseline['B_y'][5:], label='CCW', s=50, marker='o', c='g',alpha=0.2)
ax[0].scatter(baseline['C_x'][:5], baseline['C_y'][:5], label='CW', s=50, marker='x', c='b',alpha=0.2)
ax[0].scatter(baseline['C_x'][5:], baseline['C_y'][5:], label='CCW', s=50,  marker='o', c='b',alpha=0.2)

# post
ax[1].scatter(post['A_x'][:5], post['A_y'][:5], label='CW', s=50, marker='x', c='r', alpha=0.2)
ax[1].scatter(post['A_x'][5:], post['A_y'][5:], label='CCW', s=50, marker='o', c='r', alpha=0.2)
ax[1].scatter(post['B_x'][:5], post['B_y'][:5], label='CW', s=50, marker='x', c='g', alpha=0.2)
ax[1].scatter(post['B_x'][5:], post['B_y'][5:], label='CCW', s=50, marker='o', c='g', alpha=0.2)
ax[1].scatter(post['C_x'][:5], post['C_y'][:5], label='CW', s=50, marker='x', c='b', alpha=0.2)
ax[1].scatter(post['C_x'][5:], post['C_y'][5:], label='CCW', s=50, marker='o', c='b', alpha=0.2)

ax[0].set_title('Baseline')
ax[1].set_title('Post Calibration')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')

# plot mean values
ax[0].scatter(a1_x, a1_y, s=100, marker='x', c='r')
ax[0].scatter(a2_x, a2_y, s=100, marker='o', c='r')
ax[0].scatter(b1_x, b1_y, s=100, marker='x', c='g')
ax[0].scatter(b2_x, b2_y, s=100, marker='o', c='g')
ax[0].scatter(c1_x, c1_y, s=100, marker='x', c='b')
ax[0].scatter(c2_x, c2_y, s=100, marker='o', c='b')

# # plot overall mean values
# ax[0].scatter(a_x, a_y, s=100, marker='X', c='r')
# ax[0].scatter(b_x, b_y, s=100, marker='X', c='g')
# ax[0].scatter(c_x, c_y, s=100, marker='X', c='b')

# # plot mean values for post
ax[1].scatter(pa1_x, pa1_y, s=100, marker='x', c='r')
ax[1].scatter(pa2_x, pa2_y, s=100, marker='o', c='r')
ax[1].scatter(pb1_x, pb1_y, s=100, marker='x', c='g')
ax[1].scatter(pb2_x, pb2_y, s=100, marker='o', c='g')
ax[1].scatter(pc1_x, pc1_y, s=100, marker='x', c='b')
ax[1].scatter(pc2_x, pc2_y, s=100, marker='o', c='b')

# # plot overall mean values for post
# ax[1].scatter(pa_x, pa_y, s=100, marker='X', c='r')
# ax[1].scatter(pb_x, pb_y, s=100, marker='X', c='g')
# ax[1].scatter(pc_x, pc_y, s=100, marker='X', c='b')


# set x and y axes to be the same limits for both plots
ax[0].set_xlim(-25, 125)
ax[0].set_ylim(-50, 125)
ax[1].set_xlim(-25, 125)
ax[1].set_ylim(-50, 125)



plt.tight_layout()

# plot x=0, y=0
ax[0].axvline(0, color='black', lw=1, ls='--')
ax[0].axhline(0, color='black', lw=1, ls='--')
ax[1].axvline(0, color='black', lw=1, ls='--')
ax[1].axhline(0, color='black', lw=1, ls='--')

# Legend that shows all x are CW and all o are CCW
ax[0].legend(['CW', 'CCW'], loc='lower right', title='Direction')
ax[1].legend(['CW', 'CCW'], loc='lower right', title='Direction')

plt.show()


```
