```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# dummy x and y data for 3 separate groups
x11 = np.random.rand(10)*5
x12 = np.random.rand(10)
x21 = np.random.rand(10)*5
x22 = np.random.rand(10)
x31 = np.random.rand(10)*5
x32 = np.random.rand(10)

y11 = np.random.rand(10)*5
y12 = np.random.rand(10)
y21 = np.random.rand(10)*5
y22 = np.random.rand(10)
y31 = np.random.rand(10)*5
y32 = np.random.rand(10)

# create a pandas dataframe with the data
df = pd.DataFrame({
    'x': np.concatenate([x11, x21, x31]),
    'y': np.concatenate([y11, y21, y31]),
    'group': np.repeat(['group1', 'group2', 'group3'], 10),
    'cw': np.concatenate([x12, x22, x32]),
    'ccw': np.concatenate([y12, y22, y32])
})

# create a scatter plot where each group has a different marker and color and before and after are different shapes
fig, ax = plt.subplots()
groups = df.groupby('group')
for name, group in groups:
    ax.plot(group['x'], group['y'], marker='*', linestyle='', ms=12, label=name)
    ax.plot(group['cw'], group['ccw'], marker='x', linestyle='', ms=12, label=name)
ax.legend()
plt.show()


```
